```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('HollywoodMovies.csv')
```

# ANALISI 1: numero di film usciti per anno (grafico a barre)

```python
film_per_anno = df.groupby('Year')['Movie'].count()
```

```python
print(film_per_anno)
```

```python
width = 0.3
plt.bar(film_per_anno.index, film_per_anno.values, color="green", alpha = 0.5)
plt.title("Numero di film per anno")
plt.xlabel("Anno")
plt.ylabel("Numero film")
plt.show()
```

 # ANALISI 2: distribuzione 5 generi cinematografici più prodotti (Grafico a torta)

```python
movies_genre = df['Genre'].value_counts().sort_index()
```

```python
print(movies_genre)
```

```python
movies_genre = movies_genre.sort_values(ascending=False).head(5)
```

```python
print(movies_genre)
```

```python
slices = movies_genre.values
labels = movies_genre.index
colors = sns.color_palette('pastel')
plt.pie(slices, labels = labels, 
        colors=colors,
        autopct='%1.1f%%', 
        startangle=90,
        wedgeprops={'edgecolor': 'black', 'linewidth': 1.5})
plt.title('Distribuzione dei 5 generi cinematografici più prodotti')
plt.show()
```

# Analisi 3: Relazione tra budget e incassi (Grafico a dispersione)

```python
genere_budget_incassi = df.groupby('Genre')[['Budget', 'Profitability']].mean()
```

```python
print(genere_budget_incassi)
```

```python
budget_medio = genere_budget_incassi['Budget']
incassi_medi = genere_budget_incassi['Profitability']
generi = genere_budget_incassi.index
index = np.arange(14)
width = 0.3
fig, ax = plt.subplots(figsize=(12,6))
plt.bar(index, budget_medio, width=width, color= 'yellow', label='Budget Medio')
plt.bar(index+width, incassi_medi, width=width, color='pink', label='Incassi Medi' )
plt.xticks(index+width/2, generi, rotation=45)
plt.title('Confronto tra Budget Medio e Incassi Medi per Genere Cinematografico')
plt.xlabel('Genere', fontsize=12)
plt.ylabel('Importo (in milioni)', fontsize=12)
plt.legend(title="Legenda", loc='upper left')
plt.show()
```

# ANALISI 4: andamento incassi per anno (grafico a barre)

```python
profit_year = df.groupby('Year')['Profitability'].mean()
```

```python
print(profit_year)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(profit_year.index, profit_year.values, marker='o', color='blue', linestyle='-')
plt.title('Andamento degli incassi medi nel tempo')
plt.xlabel('Anno')
plt.ylabel('Incasso medio (in milioni di $)')
plt.grid(True)
plt.show()
```
